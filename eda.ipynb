```python
import pandas as pd
```

```python
df = pd.read_csv('met_museum_objects.csv')
```

```python

missing_count = df.isnull().sum()
missing_percent = df.isnull().mean() * 100

completeness_df = pd.DataFrame({
    "missing_count": missing_count,
    "missing_percent": missing_percent
}).sort_values(by="missing_percent", ascending=False)

completeness_df
```

### Completeness – Interpretation
The completeness analysis shows a strong variability across attributes.
Several fields (e.g., River, County, Excavation) exhibit more than 95% missing values, indicating structurally sparse information that is only available for specific objects.
Conversely, core descriptive and identifier attributes (e.g., Object ID, Title, Medium, Department) present high completeness, confirming the dataset’s suitability for general analysis.
This assessment guides the selection of relevant attributes for subsequent processing.


```python
#Logical Consistency Check
# Artist Begin Date ≤ Artist End Date


# Convert to numeric if needed
df["Artist Begin Date"] = pd.to_numeric(df["Artist Begin Date"], errors="coerce")
df["Artist End Date"] = pd.to_numeric(df["Artist End Date"], errors="coerce")

artist_date_inconsistency = df[
    (df["Artist Begin Date"].notna()) &
    (df["Artist End Date"].notna()) &
    (df["Artist Begin Date"] > df["Artist End Date"])
]

len(artist_date_inconsistency)
```

```python
#Logical Consistency Check
# Object Begin Date ≤ Object End Date

df["Object Begin Date"] = pd.to_numeric(df["Object Begin Date"], errors="coerce")
df["Object End Date"] = pd.to_numeric(df["Object End Date"], errors="coerce")

object_date_inconsistency = df[
    (df["Object Begin Date"].notna()) &
    (df["Object End Date"].notna()) &
    (df["Object Begin Date"] > df["Object End Date"])
]

len(object_date_inconsistency)
```

```python

artist_role_inconsistency = df[
    df["Artist Display Name"].isna() &
    df["Artist Role"].notna()
]

len(artist_role_inconsistency)
```

```python
artist_info_inconsistency = df[
    df["Artist Display Name"].isna() & (
        df["Artist Begin Date"].notna() |
        df["Artist End Date"].notna() |
        df["Artist Nationality"].notna()
    )
]

len(artist_info_inconsistency)
```

```python
artist_object_time_inconsistency = df[
    df["Artist Begin Date"].notna() &
    df["Object Begin Date"].notna() &
    (df["Object Begin Date"] < df["Artist Begin Date"])
]

len(artist_object_time_inconsistency)
```

```python
public_domain_inconsistency = df[
    (df["Is Public Domain"] == True) &
    (df["Rights and Reproduction"].notna())
]

len(public_domain_inconsistency)
```

```python
geo_inconsistency = df[
    df["City"].notna() & df["Country"].isna()
]

len(geo_inconsistency)
```

```python
# Classification value counts
class_counts = df["Classification"].value_counts(dropna=False)
class_percent = df["Classification"].value_counts(normalize=True, dropna=False) * 100

classification_summary = pd.DataFrame({
    "count": class_counts,
    "percent": class_percent
})

classification_summary
```
